```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

path_py = 'colors_py/'


##### formation redshift 70 => zf=7.0
zf=70

file_name = 'couleurs_yjhbk_zf_'+str(zf)+'.dat'

###### read the file using numpy.loadtxt

data = np.loadtxt(path_py+file_name).T

#### extract the u, g, r, i and z bands from the model. The redshifts are in the last element
mod_u = data[0]
mod_g = data[1]
mod_r = data[2]
mod_i = data[3]
mod_z = data[4]
mod_redshift = data[5]


#### plot colors from the model for all redshfts

plt.figure(figsize=(15,10))
plt.plot(mod_redshift, mod_g - mod_r, lw=4, label='g-r')
plt.plot(mod_redshift, mod_g - mod_i, lw=4, label='g-i')
plt.plot(mod_redshift, mod_g - mod_u, lw=4, label='g-u')
plt.plot(mod_redshift, mod_g - mod_z, lw=4, label='g-z')
plt.xlabel('Redshift')
plt.ylabel('Colors')
plt.legend()
plt.show()


```
